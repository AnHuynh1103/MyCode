```python
#Regression #Hồi quy: tìm ra 1 mối quan hệ giữa 2 hoặc nhiều biến
#predict tiên đoán

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Salary_Data.csv')
data.head(30)
```

```python
data.shape
```

```python
data[data['YearsExperience']==2.2].head()
```

```python
x=data.drop(['Salary'],axis=1)
x
```

```python
y=data.drop(['YearsExperience'],axis=1)
y
```

```python
sns.pairplot(data)
```

```python
#Vẽ đồ thị
plt.plot(x,y,'r',label='Data')
```

```python
plt.plot(x,y,'r.',label='Data')
```

```python
plt.scatter(x,y,c='r',label='Data')
```

```python
plt.scatter(x,y,c='r',label='Data')
plt.title('Years Experience and Salary')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y) #fit dạy cho máy học
print('Giá trị intercpet là:',reg.intercept_)
print('Phương trình hồi quy là: y = ',reg.coef_,'x +',reg.intercept_)
```

```python
y_pred = reg.predict(x).round(1)
cal = pd.DataFrame(np.c_[y,y_pred],columns=['Original Salary','Predicted Salary'])
cal
```

```python
#Tạo dữ liệu mới
error = y - y_pred
cal_new = pd.DataFrame(np.c_[x,y,y_pred,error],columns=['Years Experence','Original Salary','Predicted Salary','Error'])
cal_new
```

```python
reg.score(x,y) #khoảng 0.95 <=> 95%
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,y_pred,c='b')
plt.title('Salary and Years Experence')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_new = pd.read_csv('1.01. Simple linear regression.csv')
data_new.head(84)
```

```python
data_new.shape
```

```python
a=data_new.drop(['GPA'],axis=1)
a
```

```python
b=data_new.drop(['SAT'],axis=1)
b
```

```python
sns.pairplot(data_new)
```

```python
plt.plot(a,b,'r',label='Data new')
```

```python
plt.scatter(a,b,c='r',label='Data')
```

```python
from sklearn.linear_model import LinearRegression
reg_new = LinearRegression()
reg_new.fit(a,b) #fit dạy cho máy học
print('Giá trị intercpet là:',reg_new.intercept_)
print('Phương trình hồi quy là: y = ',reg_new.coef_,'x +',reg_new.intercept_)
```

```python
b_pred = reg_new.predict(a).round(1)
cal_new_1 = pd.DataFrame(np.c_[b,b_pred],columns=['Original GPA','Predicted GPA'])
cal_new_1
```

```python
error1 = b - b_pred
cal_new_2 = pd.DataFrame(np.c_[a,b,b_pred,error1],columns=['SAT','Original GPA','Predicted GPA','Error'])
cal_new_2
```

```python
reg_new.score(a,b)
```

```python
plt.figure(figsize=(9,9))
plt.scatter(a,b,c='r')
plt.plot(a,b_pred,c='b')
plt.title('SAT and GPA')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```
