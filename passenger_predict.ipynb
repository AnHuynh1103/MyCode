```python
import pandas as pd
import numpy as np
url = 'passenger1.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
data.head(10)
```

```python
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='red')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Miles(X)']),y=data['Dollars (Y)'])
print(model.intercept_)
print(model.coef_)

#The accuracy of the first order
print('The value of intercpet is:',model.intercept_)
print('The regression equation is: y = ',model.coef_,'x +',model.intercept_)
print('The value of model accuracy is:',model.score(X=pd.DataFrame(data['Miles(X)']),y=data['Dollars (Y)'])*100,'%')
```

```python
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='red',xlim=(0,6000))
y_pred = model.predict(X=pd.DataFrame(data['Miles(X)']))
plt.plot(data['Miles(X)'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression()
X2 = pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2]).T
model.fit(X2,y=data['Dollars (Y)'])

#The accuracy of the second order
print('The value of intercpet is:',model.intercept_)
print('The values of Coeffient is: ',model.coef_)
print('The value of model accuracy is:',model.score(X2,y=data['Dollars (Y)'])*100,'%')
```

```python
range = np.arange(300,5900,0.01)
data.plot(kind='scatter',x='Miles(X)',y='Dollars (Y)',figsize=(9,9),color='red',xlim=(0,6000))
predictor = pd.DataFrame([range,range**2]).T
y_predict = model.predict(X=predictor)
plt.plot(range,y_predict,color='blue')
```
