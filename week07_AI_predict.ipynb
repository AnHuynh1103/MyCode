```python
import pandas as pd
import numpy as np
url = 'mtcars.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print(model.intercept_)
print(model.coef_)

print('Giá trị intercpet là:',model.intercept_)
print('Phương trình hồi quy là: y = ',model.coef_,'x +',model.intercept_)
```

```python
model.score(X=pd.DataFrame(data['wt']),y=data['mpg'])
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='red',xlim=(0,7))
y_pred = model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
correlation=data.corr()
print(correlation)
```

```python
x = data.drop(['radio','newspaper','sales'],axis=1)
x
```

```python
y = data.drop(['radio','newspaper','TV'],axis=1)
y
```

```python
plt.scatter(x,y,c='black',label='Data')
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y) #fit dạy cho máy học
print('Giá trị intercpet là:',reg.intercept_)
print('Phương trình hồi quy là: y = ',reg.coef_,'x +',reg.intercept_)
print('Độ chính xác của mô hình là: ',reg.score(x,y))
```

```python
y_pred = reg.predict(x).round(1)
cal = pd.DataFrame(np.c_[y,y_pred],columns=['Original sales','Predicted sales'])
cal
```
